```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.scatter(X, y, color='blue')
plt.title("Synthetic Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
w = 0.0  # weight
b = 0.0  # bias
learning_rate = 0.01
epochs = 100
```

```python
m = len(X)
```

```python
for epoch in range(epochs):
    y_pred = w * X + b
    error = y_pred - y

    # Calculate gradients
    dw = (2/m) * np.sum(error * X)
    db = (2/m) * np.sum(error)

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    if epoch % 10 == 0:
        loss = np.mean(error ** 2)
        print(f"Epoch {epoch}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, w * X + b, color='red', label='Regression Line')
plt.title("Linear Regression Result")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
